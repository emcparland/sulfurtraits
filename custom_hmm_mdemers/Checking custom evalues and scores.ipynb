```r
library(dplyr)
library(tidyr)
library(data.table)
library(ggplot2)
```

```r
txt <- list.files(pattern = ".hmm.txt")
txt = txt[-which(txt == "allcustom_filtered.hmm.txt")]
gene.df <- data.frame(txt.files = txt, gene = gsub(txt, pattern = ".hmm.txt", replacement = ""), 
           original.cutoff = c(175,410,100,271,150,188,130,382,223,91,411,682,202,550,400,200,400,100,266,400,374,258))
```

```r
gene.df$new_filter = NA

for(i in 1:dim(gene.df)[1]){

    gene = gsub(gene.df$txt.files[i], pattern = ".hmm.txt", replacement = "")

    p <- fread(txt[i],header = F) %>%
    mutate(target = V1, query = V3, full_evalue = V5, full_score = V6) %>%
    select(c(target,query,full_evalue,full_score))
    
    gene.df$new_filter[i] = as.numeric(summary(p$full_score)[5])
    
    p <- p %>%
    pivot_longer(cols = c('full_evalue', 'full_score')) %>%
    filter(name != "full_evalue") %>%
    #filter(value > gene.df$original.cutoff[i]) %>%
    ggplot(aes(x = value))+
    geom_histogram()+
    geom_vline(xintercept = gene.df$original.cutoff[i])+
    geom_vline(xintercept = gene.df$new_filter[i])+
    facet_wrap(~name, scales = "free")+
    ggtitle(gene)
    theme_bw()+theme(text = element_text(size = 16))
    print(p)
}
```

```r
write.table(gene.df, file = "customhmm_scorecutoffs.txt", quote = F, sep = "\t", row.names = F)
```
